```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Word-embedding.csv')
```

```python
df.head()
```

```python
DL = df.iloc[:6]
DL
```

```python
bg_color = (.1,.40,.60)
sns.set(rc={"font.style":"normal",
            "axes.facecolor":bg_color,
            "figure.facecolor":bg_color,
            "text.color":"white",
            "xtick.color":"white",
            "ytick.color":"white",
            "axes.labelcolor":"white",
            "axes.grid":False,
            'axes.labelsize':30,
            'figure.figsize':(15.0, 7.0),
            'xtick.labelsize':25,
            'ytick.labelsize':30.0})
plt.Figure(figsize = (8,20))
f = sns.barplot( x = DL['Algorithm'], y = DL['Accuracy'], data = DL,edgecolor=(0,1,1),
                  linewidth=2,
                  palette="Accent_r")
f.set_xlabel(' Deep Learning Algorithm')
f.set_ylabel("Accuracy")
#f = g.legend(["No,","Yes"],fontsize = 30)
f.set_xticklabels(f.get_xticklabels(),rotation=20)
plt.title('Accuracy comparision among the Deep Learning Algorithm', fontsize = 40)
for rect in f.patches:
    f.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.005,rect.get_height(),horizontalalignment='center', fontsize = 25.0)
   
```
